# Characteristics_Micorbiome_Plot.ipynb
The following code is used to visualize the results generated by step3-02-Characteristics_Micorbiome

```python
import matplotlib.pyplot as plt
import pickle
```

```python
#read_data
def read_data(data_pkl,label_pkl,samples_pkl):
    with open(data_pkl,'rb') as f:
        data = pickle.load(f)
    with open(label_pkl,'rb') as f:
        label = pickle.load(f)
    with open(samples_pkl,'rb') as f:
        samples = pickle.load(f)
    return data,label,samples
```

```python
#core
def plotboxes_reverse(data,labels,figsize):
    data = data[:10]
    labels = labels[:10]
    data.reverse()
    labels.reverse()
    y_position = [5]
    for i in range(len(data)-1): # boxes_position  and x axis position
        y_position.append(y_position[-1]+10)
    plt.figure(figsize=figsize)
    #plt.grid(True)  # 演示网格
    bplot1= plt.boxplot(data,
                patch_artist=True,
                boxprops=dict(facecolor="LightSeaGreen"),
                medianprops={'color': 'black', 'linewidth': '1'},
                meanline=True,
                showmeans=True,
                meanprops={'color': 'blue', 'ls': '-', 'linewidth': '1'},
                flierprops={"marker": "o", "markerfacecolor": "green", "markersize": 3},
                widths = 7.5,
                positions=y_position,
                       vert = False)
    plt.yticks([i for i in y_position], labels,fontsize= 10,ha='right') # 旋转90度
    plt.xticks(fontsize= 10)
    plt.xlabel('log2(T/N)',fontsize=12)
    plt.title('Bacteria_16sRna')
    #plt.savefig("D:/ct/Bioinformatics/微生物/new_microbial/results/16s_rna/03-characteristic_bacteria/"+'cb_species.png',dpi=300,bbox_inches='tight')
    plt.show()
data_pkl = "../step3-select_high_abundance_and_characteristics_micorbiome/results/characteristics_micorbiome/data.pkl"
labels_pkl = "../step3-select_high_abundance_and_characteristics_micorbiome/results/characteristics_micorbiome/labels.pkl"
samples_pkl = "../step3-select_high_abundance_and_characteristics_micorbiome/results//characteristics_micorbiome/samples.pkl"
data,labels,samples = read_data(data_pkl,labels_pkl,samples_pkl)
plotboxes_reverse(data,labels,(5,5))
```
